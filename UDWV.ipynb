# **Univariate Plots**

**Histograms**

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = '/content/sample_data/pima-indians-diabetes.data.csv'

# Load dataset assuming the CSV file has headers
data = pd.read_csv(filename)

# Plot histograms for each feature
data.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle("Univariate Histograms")
plt.tight_layout()
plt.show()
```

# **Box and Whisker Plots**

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = "/content/sample_data/pima-indians-diabetes.data.csv"

# Try reading with no custom headers first
try:
    data = pd.read_csv(filename)
except pd.errors.ParserError:
    # If the file has no headers, fall back
    column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
    data = pd.read_csv(filename, names=column_names, header=None)

# Convert all columns to numeric (safeguard against string parsing issues)
data = data.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaNs caused by conversion errors (optional)
data.dropna(inplace=True)

# Plot Box and Whisker Plots
data.plot(kind='box', subplots=True, layout=(3, 3), sharex=False, sharey=False, figsize=(10, 8))
plt.suptitle("Box and Whisker Plots")
plt.tight_layout()
plt.show()
```

# **Multivariate Plots**

**Correlation Matrix Plot**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filename = '/content/sample_data/pima-indians-diabetes.data.csv'
column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# Load the data assuming no header in CSV file
data = pd.read_csv(filename, names=column_names, header=None)

# Ensure data is numeric
data = data.apply(pd.to_numeric, errors='coerce')
data.dropna(inplace=True)

# Compute correlation matrix
correlations = data.corr()

# Plot correlation matrix
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap='coolwarm')
fig.colorbar(cax)

# Tick marks and labels
ticks = np.arange(len(column_names))
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(column_names, rotation=45, ha="left")
ax.set_yticklabels(column_names)

plt.title("Correlation Matrix", pad=20)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = '/content/sample_data/pima-indians-diabetes.data.csv'
column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# Load dataset assuming no header
data = pd.read_csv(filename, names=column_names, header=None)

# Convert to numeric, drop NaNs
data = data.apply(pd.to_numeric, errors='coerce')
data.dropna(inplace=True)

# Compute correlation matrix
correlations = data.corr()

# Plot correlation matrix
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap='coolwarm')
fig.colorbar(cax)

# Tick labels
ax.set_xticks(range(len(column_names)))
ax.set_yticks(range(len(column_names)))
ax.set_xticklabels(column_names, rotation=45, ha='left')
ax.set_yticklabels(column_names)

plt.title("Correlation Matrix", pad=20)
plt.tight_layout()
plt.show()
```

# **Scatter Plot Matrix**

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

filename = "/content/sample_data/pima-indians-diabetes.data.csv"
column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# Load dataset assuming no header row
data = pd.read_csv(filename, names=column_names, header=None)

# Ensure all data is numeric
data = data.apply(pd.to_numeric, errors='coerce')
data.dropna(inplace=True)

# Create scatterplot matrix
scatter_matrix(data, figsize=(12, 10), diagonal='kde')
plt.suptitle("Scatterplot Matrix")
plt.tight_layout()
plt.show()
```
